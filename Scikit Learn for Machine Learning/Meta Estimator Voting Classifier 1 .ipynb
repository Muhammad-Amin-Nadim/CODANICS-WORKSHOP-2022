```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=2000,n_features=2,n_redundant=0,random_state=21)
plt.scatter(x=X[:,0],y=X[:,1],c=y,s=25,edgecolors='black')
```

```python
clf1 = LogisticRegression().fit(X,y)
clf2=KNeighborsClassifier().fit(X,y)
clf3=DecisionTreeClassifier().fit(X,y)
clf4=VotingClassifier(
    estimators=[("LogisticRegression:",clf1),
                ("KNN:",clf2),
                ("Decision Tree:",clf3)],
                voting='hard'
)
```

```python
clf1.predict(X)
clf1.score(X,y)
```

```python
clf2.predict(X)
clf2.score(X,y)
```

```python
clf3.predict(X)
clf3.score(X,y)
```

```python
clf4.fit(X,y)
clf4.predict(X)
clf4.score(X,y)
```
